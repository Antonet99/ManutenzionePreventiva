```python
import pandas as pd
import os
```

```python
data = pd.read_csv('train/data/Case001.csv')
data.head()
```

```python
dfs = []

for i in range(2, 178):
    file_path = f'train/data/Case{i:03d}.csv'
    if os.path.exists(file_path):
        df = pd.read_csv(file_path)
        dfs.append(df)

# Concatenate the dataframes
concatenated_df = pd.concat([data] + dfs, ignore_index=True)

# Print the concatenated dataframe
concatenated_df
```

```python
concatenated_df.to_csv('concatenated.csv', index=False)
```
